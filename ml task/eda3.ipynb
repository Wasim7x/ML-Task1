EDA ON SEA LEVEL

GMSL --->>> global mean sea level

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

crbon emision start

```python
df = pd.read_csv('co2.csv')
df
```

```python
df = df[(df['Year'] >= 1993) & (df['Year'] <= 2015)]
```

```python
df = df[['Country', 'CO2 emission (Tons)', 'Year',]]
df
```

```python
df1 = df.groupby('Year').agg({"CO2 emission (Tons)": "median"})
df1
```

```python
df1 = df1.reset_index()
df1
```

carbon emission end

sea level start

```python
df = pd.read_csv('sealevel.csv')
df
```

```python
type(df['Time'])
```

```python
df['Time'] = pd.to_datetime(df['Time']) 
```

```python
df['Year'] = df['Time'].dt.year
df['Year'] 
```

```python
df2 = df.groupby('Year')['GMSL'].median().reset_index()
df2
```

```python
df2.dtypes
```

sea level end

```python

plt.plot(df2['Year'], df2['GMSL'], label='Sea level', color='r')
plt.xlabel("Year")
plt.ylabel('GMSL')
plt.title('GMSL per year')
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.plot(df1['Year'], df1['CO2 emission (Tons)'], label='CO2', color='g')
plt.xlabel("Year")
plt.ylabel('CO2 emission (Tons)')
plt.title('CO2 emission')
plt.grid(True)
plt.legend()
plt.show()
```
