```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('co2.csv')
```

```python
df
```

```python
df['co2 emition per Area(km2)']=df['CO2 emission (Tons)']/df['Area']
df['co2 emition per Area(km2)']
```

```python
df2 = df.groupby('Year').agg({"co2 emition per Area(km2)": "median"})
df3 = df[df['Country'] == 'India'].groupby('Year').agg({"co2 emition per Area(km2)": "median"})
df3
```

```python
plt.plot(df2['co2 emition per Area(km2)'].keys(), df2['co2 emition per Area(km2)'].values,label= 'world', color='r')
plt.plot(df3['co2 emition per Area(km2)'].keys(), df3['co2 emition per Area(km2)'].values,label= 'india',color='g')
plt.xlabel("Year")
plt.ylabel('Avg carbon emission')
plt.title('World Avg Carbon Emission(km2))')
plt.grid(True)
plt.legend()
plt.show()
```

```python
df2 = df.groupby('Year').agg({"co2 emition per Area(km2)": "median"})
df4 = df[df['Country'] == 'Russia'].groupby('Year').agg({"co2 emition per Area(km2)": "median"})
df5 = df[df['Country'] == 'United States'].groupby('Year').agg({"co2 emition per Area(km2)": "median"})
```

```python
plt.plot(df2['co2 emition per Area(km2)'].keys(), df2['co2 emition per Area(km2)'].values,label= 'world', color='r')
plt.plot(df4['co2 emition per Area(km2)'].keys(), df4['co2 emition per Area(km2)'].values,label= 'Russia', color='b')
plt.plot(df5['co2 emition per Area(km2)'].keys(), df5['co2 emition per Area(km2)'].values,label= 'U.S.A', color='g')
plt.xlabel("Year")
plt.ylabel('Avg carbon emission')

plt.title('World Avg Carbon Emission(per person)')
plt.grid(True)
plt.legend()
plt.show()
```
